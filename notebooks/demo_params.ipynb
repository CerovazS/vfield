```python
from vfield import render_flow_field
import torch
import matplotlib.pyplot as plt

vfield = torch.load('../data/vectorfield.pt')

grid = vfield['grid']
displacement=vfield['displacement']
```

```python
plt.figure(figsize=(10, 5))

for i, steps in enumerate([10, 25, 100]):

    ff_img = render_flow_field(grid, displacement, steps=steps)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, origin='lower')
    plt.title(f'steps={steps}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

for i, step_size in enumerate([0.05, 0.01, 0.002]):

    ff_img = render_flow_field(grid, displacement, step_size=step_size)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, origin='lower')
    plt.title(f'step_size={step_size}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

for i, particles in enumerate([100, 1000, 10_000]):

    ff_img = render_flow_field(grid, displacement, particles=particles)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, origin='lower')
    plt.title(f'particles={particles}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(render_flow_field(grid, displacement, antialias=False), origin='lower')
plt.title('no antialias')
plt.axis('off')

for i, aa_factor in enumerate([2, 4]):

    ff_img = render_flow_field(grid, displacement, antialias=True, aa_factor=aa_factor)

    plt.subplot(1, 3, i+2)
    plt.imshow(ff_img, origin='lower')
    plt.title(f'aa_factor={aa_factor}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

bg_colors = [(255, 255, 255), (0, 0, 0), (144, 210, 206)]
trace_colors = [(0, 0, 0), (255, 255, 255), (73, 101, 123)]

for i, (bg, fg) in enumerate(zip(bg_colors, trace_colors)):

    ff_img = render_flow_field(grid, displacement, bg_color=bg, trace_color=fg)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, origin='lower')
    plt.title(f'color scheme {i+1}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
