```python
from vfield import render_flow_field
import torch
import matplotlib.pyplot as plt

vfield = torch.load('data/vectorfield.pt')

grid = vfield['grid']
displacement=vfield['displacement']
```

```python
plt.figure(figsize=(10, 5))

for i, steps in enumerate([10, 25, 100]):

    ff_img = render_flow_field(grid, displacement, steps=steps)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, cmap='PuBu', origin='lower')
    plt.title(f'steps={steps}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

for i, step_size in enumerate([0.05, 0.01, 0.002]):

    ff_img = render_flow_field(grid, displacement, step_size=step_size)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, cmap='PuBu', origin='lower')
    plt.title(f'step_size={step_size}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

for i, particles in enumerate([100, 1000, 10_000]):

    ff_img = render_flow_field(grid, displacement, particles=particles)

    plt.subplot(1, 3, i+1)
    plt.imshow(ff_img, cmap='PuBu', origin='lower')
    plt.title(f'particles={particles}')
    plt.axis('off')

plt.tight_layout()
plt.show()
```
